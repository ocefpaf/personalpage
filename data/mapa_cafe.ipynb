```python
from folium.folium import Map
from IPython.display import HTML


def inline_map(m):
    """Takes a folium instance or a html path and load into an iframe."""
    if isinstance(m, Map):
        m._build_map()
        srcdoc = m.HTML.replace('"', '&quot;')
        embed = HTML('<iframe srcdoc="{srcdoc}" '
                     'style="width: 100%; height: 500px; '
                     'border: none"></iframe>'.format(srcdoc=srcdoc))
    elif isinstance(m, str):
        embed = HTML('<iframe src="{src}" '
                     'style="width: 100%; height: 500px; '
                     'border: none"></iframe>'.format(src=m))
    return embed
```

```python
import numpy as np


coffees = dict(catuai_vermelho=dict(local="Sitio Corrego da Prata.  Castelo, ES.",
                                    pos=(-41-20/60. - 17/60/60., -20-59/60. -35/60/60.),
                                    aromas="frutas amarelas e caramelo",
                                    corpo="alto e macio",
                                    docura="alta, melado de cana",
                                    obs=u"Não tem personalidade, mas quando esfria aumento o sabor.  Se redimiu."),
               catuai_amarelo=dict(local="Sitio Rio Dantas.  Gleba do Paraiso, MG",
                                   pos=(-46-36/60. -52/60/60., -21-44/60. -14/60/60.),
                                   aromas="frutas tropicais, floral e chocolate",
                                   corpo=u"lácteo e aveludado",
                                   docura="alta",
                                   obs=u"+ gostoso que o CV, bom, mas falta algo q tem no BV!"),
               bourbon_vermelho=dict(local="Fazenda Bau.  Lagoa Formosa, MG",
                                     pos=(-47-3/60.-38.06/60/60., -22-54/60.-21.64/60/60.),
                                     aromas="frutas vermelhas e castanhas",
                                     corpo=u"alto",
                                     docura="alta",
                                     obs=u"... and the oscar goes to !!")
              )

positions = []
for coffee, data in coffees.items():
    positions.append(data['pos'])

positions = np.array(positions)
lon_center, lat_center = positions.mean(axis=0)
```

```python
import folium

coffee_map = folium.Map(location=[lat_center, lon_center], zoom_start=5)

for coffee, data in coffees.items():
    popup = '<b>{}</b>'.format(data['local'])
    kw = dict(popup=popup, marker_color="green", marker_icon="ok")
    coffee_map.simple_marker(location=[data['pos'][1], data['pos'][0]], **kw)

coffee_map.create_map(path='coffee_map.html')
coffee_map.render_iframe = True
inline_map(coffee_map)
```
